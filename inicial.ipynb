```python
import pandas as pd

tabela = pd.read_csv("clientes.csv")
display(tabela)
```

```python
# Verificar se a base tem problemas
display(tabela.info())

# Good - Boa
# Standard - MÃ©dia
#Poor - Ruim

# float - numeros com casa decimal
# object - texto
# int - numeros inteiros
```

```python
# Colunas que sÃ£o texto: (tratamento)
    # profissÃ£o
    # mix_crÃ©dito
    # comportamento_pagamento

# Label Incoder
from sklearn.preprocessing import LabelEncoder

codifcador = LabelEncoder()

#tabela["profissao"] = codifcador.fit_transform(tabela["profissao"]) A tabela["profissao"] serÃ¡ transformada em nÃºmeros pelo codificador (codificaÃ§Ã£o das profissÃµes)

tabela["profissao"] = codifcador.fit_transform(tabela["profissao"])

tabela["mix_credito"] = codifcador.fit_transform(tabela["mix_credito"])

tabela["comportamento_pagamento"] = codifcador.fit_transform(tabela["comportamento_pagamento"])

display(tabela.info())
```

```python
# x Ã© quem a IA pode usar pra fazer previsÃ£o
# y Ã© quem a IA tem que prever

#x = tabela.drop(columns=["score_credito", "id_cliente"])
#y = ["score_credito"]

#from sklearn.model_selection import train_test_split

#x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)

x = tabela.drop(["score_credito", "id_cliente"], axis=1)
y = tabela["score_credito"]

from sklearn.model_selection import train_test_split

# separamos os dados em treino e teste. Treino vamos dar para os modelos aprenderem e teste vamos usar para ver se o modelo aprendeu corretamente
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
# Modelos
# Ãrvore de decisÃ£o -> RandonForest
# Neirest Neighbors -> KNN -> Vizinhos PrÃ³ximos

# Importar a IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# Criar a IA
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# Treinar a IA
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)

# Base de dados para treinar -> Kaggle
```

```python
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

from sklearn.metrics import accuracy_score

display(accuracy_score(y_teste, previsao_arvoredecisao))
display(accuracy_score(y_teste, previsao_knn))
```

```python
# Pra esse desafio o modelo da Ã¡rvore de decisÃ£o Ã©  melhor modelo
# Usar o nosso modelo pra fazer novas previsÃµes

# Importar novos clientes pra prever
tabela_novos_clientes = pd.read_csv("novos_clientes.csv")
# display(tabela_novos_clientes) para ver as profissÃµes dos clientes em texto (object)

# Ajustar as colunas de texto
tabela_novos_clientes["profissao"] = codifcador.fit_transform(tabela_novos_clientes["profissao"])

tabela_novos_clientes["mix_credito"] = codifcador.fit_transform(tabela_novos_clientes["mix_credito"])

tabela_novos_clientes["comportamento_pagamento"] = codifcador.fit_transform(tabela_novos_clientes["comportamento_pagamento"])

# Fazer previsÃµes
previsoes = modelo_arvoredecisao.predict(tabela_novos_clientes)

display(tabela_novos_clientes)
display(previsoes)
```
